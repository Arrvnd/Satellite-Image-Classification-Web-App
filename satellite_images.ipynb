```python
import os
import matplotlib.pyplot as plt
import cv2
import numpy as np
import pandas as pd
from PIL import Image
```

```python
folder_path = r"C:\Users\aravi\satellite\cloudy"
image_num_data = list()
for file_name in os.listdir(folder_path):
        img_path = os.path.join(folder_path, file_name)
        img = Image.open(img_path).convert('RGB')
        img = np.array(img)
        img = cv2.resize(src = img, dsize = (224,224),interpolation=cv2.INTER_LINEAR)
        image_num = np.array(img)
        image_flat = image_num.flatten()
        image_num_data.append(image_flat)
image_array = np.array(image_num_data)
df1 = pd.DataFrame(image_array)
df1['type'] = 'cloudy'
df1
```

```python
folder_path = r"C:\Users\aravi\satellite\desert"
image_num_data = list()
for file_name in os.listdir(folder_path):
        img_path = os.path.join(folder_path, file_name)
        img = Image.open(img_path).convert('RGB')
        img = np.array(img)
        img = cv2.resize(src = img, dsize = (224,224),interpolation=cv2.INTER_LINEAR)
        image_num = np.array(img)
        image_flat = image_num.flatten()
        image_num_data.append(image_flat)
image_array = np.array(image_num_data)
df2 = pd.DataFrame(image_array)
df2['type'] = 'desert'
df2
```

```python
folder_path = r"C:\Users\aravi\satellite\green_area"
image_num_data = list()
for file_name in os.listdir(folder_path):
        img_path = os.path.join(folder_path, file_name)
        img = Image.open(img_path).convert('RGB')
        img = np.array(img)
        img = cv2.resize(src = img, dsize = (224,224),interpolation=cv2.INTER_LINEAR)
        image_num = np.array(img)
        image_flat = image_num.flatten()
        image_num_data.append(image_flat)
image_array = np.array(image_num_data)
df3 = pd.DataFrame(image_array)
df3['type'] = 'green'
df3
```

```python
folder_path = r"C:\Users\aravi\satellite\water"
image_num_data = list()
for file_name in os.listdir(folder_path):
        img_path = os.path.join(folder_path, file_name)
        img = Image.open(img_path).convert('RGB')
        img = np.array(img)
        img = cv2.resize(src = img, dsize = (224,224),interpolation=cv2.INTER_LINEAR)
        image_num = np.array(img)
        image_flat = image_num.flatten()
        image_num_data.append(image_flat)
image_array = np.array(image_num_data)
df4 = pd.DataFrame(image_array)
df4['type'] = 'water'
df4
```

```python
df = pd.concat([df1,df2,df3,df4],ignore_index = True)
df = df.sample(frac = 1).reset_index(drop = True)
df
```

```python
y = df['type'] 
X = df.drop('type',axis = 1)
```

```python
from sklearn.decomposition import PCA
pca_new = PCA(n_components=500)
pca_new_data = pca_new.fit_transform(X)
print("shape = ",pca_new_data.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(pca_new_data,y,test_size = 0.2,random_state = 42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_encoded = sc.fit_transform(X_train)
X_test_encoded = sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier()
rf_model.fit(X_train_encoded,y_train)
```

```python
y_pred = rf_model.predict(X_test_encoded)
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test,y_pred)
score
```
